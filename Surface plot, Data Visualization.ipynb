# Surface Plot | Data Visualization

### Surface plots used for :
    - Visualise Loss functions in ML & Deep Learning
    - Visualise state or state value function in Reinforcement Learning

```python
import numpy as np
import matplotlib.pyplot as plt
# from mpl_toolkits.mpplot3d import Axes3D
```

```python
a= np.array([1,2,3])
b= np.array([4,5,6,7])

a,b = np.meshgrid(a,b)
print(a)
print(b)
```

```python
fig = plt.figure()
axes = fig.gca(projection='3d')
axes.plot_surface(a,b,a**2+b**2,cmap='rainbow')
plt.show()
```

```python
x= np.arange(-1,1,0.02)
y=x
x,y=np.meshgrid(x,y)
fig=plt.figure()
axes=fig.gca(projection='3d')
axes.plot_surface(x,y,x**2+y**2,cmap='rainbow')
plt.show()
```

```python
fig=plt.figure()
axes=fig.gca(projection='3d')
axes.contour(x,y,x**2+y**2,cmap='rainbow')
plt.title('contour plot')
plt.show()
```
