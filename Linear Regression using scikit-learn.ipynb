```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import make_regression
%matplotlib notebook
```

```python
X,y = make_regression(n_samples=1000,n_features=2, n_informative=2, noise=1,random_state=1)
print(X.shape, y.shape)
```

```python
plt.subplot(1,2,1)
plt.scatter(X[:,0],y)

plt.subplot(1,2,2)
plt.scatter(X[:,1],y)
```

```python
#3d plot
from mpl_toolkits import mplot3d

fig = plt.figure(figsize=(10,7))
ax = plt.axes(projection = "3d")

#creating plot
ax.scatter3D(X[:,0], X[:,1], y, color = "green")
plt.title("3D scatter plot")
plt.show()
```

```python
#implementation of regression of model
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(X,y)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
model.predict([X[0],X[1]]) #these are predicted output of y0, y1
```

```python
#and actual y0, y1 is
y[0], y[1]
```

```python
#score of the model by giving input a train data
model.score(X,y) #it's about 99% 
```
